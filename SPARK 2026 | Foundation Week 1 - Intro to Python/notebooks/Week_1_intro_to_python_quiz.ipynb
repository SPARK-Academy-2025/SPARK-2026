## ‚ùì Question 19 ‚Äî Counting with Loops

```python
results = ["Normal", "Abnormal", "Normal", "Abnormal", "Normal"]

count = 0
for r in results:
    if r == "Abnormal":
        count += 1

print(f"Abnormal scans: {count}")
```
