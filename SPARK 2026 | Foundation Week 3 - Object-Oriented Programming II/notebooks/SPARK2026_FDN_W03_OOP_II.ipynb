## 2️⃣ Python Inheritance  

**Inheritance** allows us to define a class that **inherits all methods and properties** from another class.  

### 🔹 Key Concepts:  
- 🏗 **Parent Class** – The class being inherited from (*also called the base class*).  
- 🏗 **Child Class** – The class that inherits from another class (*also called the derived class*).  

Inheritance promotes **code reusability** and allows for **hierarchical class structures**. 🚀  


## 2.1️⃣ *Creating a Parent Class*  

We can create a **Parent Class** that contains properties and methods which can be inherited by other classes.  

### ✨ Example: Defining a `Patient` Class  


````python
class Patient:
    def __init__(self, bmi, health_condition):
        self.bmi = bmi
        self.health_condition = health_condition

    def print_info(self):
        print(f"Patient BMI: {self.bmi}, Health Condition: {self.health_condition}")

# Create an object of the Patient class
patient1 = Patient(24.5, "Healthy")
patient1.print_info()
````

## 2.2️⃣ *Creating a Child Class*  

A **Child Class** inherits properties and methods from a **Parent Class**.  
To create a child class, pass the parent class as a parameter when defining it.  

### ✨ Example: Defining an `EmergencyPatient` Class 

````python
class EmergencyPatient(Patient):
    def __init__(self, bmi, health_condition, severity_level):
        super().__init__(bmi, health_condition)
        self.severity_level = severity_level

    def print_emergency_info(self):
        print(f"Patient BMI: {self.bmi}, Health Condition: {self.health_condition}, Severity Level: {self.severity_level}")

# Create an object of the EmergencyPatient class
patient2 = EmergencyPatient(30.2, "At Risk", "High")
patient2.print_emergency_info()
````

### 🔹 Explanation:
- 🏗 **`EmergencyPatient` inherits from `Patient`** using `super().__init__()`.  
- 🔄 **Adds a new property `severity_level`** to indicate emergency level.  
- 📤 **Includes a `print_emergency_info()` method** to display extended patient details.  
- ✅ **Creates an object `patient2`** and calls the method to print information.  
- 🚀 The **child class extends** the functionality of the **Parent Class** while reusing its properties.

### ✨ Example: Using `super()` in `EmergencyPatient` 

# 🔗 **Inheritance in Python**  

Inheritance is a **core concept** in Object-Oriented Programming (**OOP**) that enables a class to **inherit properties and behaviors** from another class.  

### ✨ Key Concepts:  
- 🏗 **Parent/Base Class** – The existing class whose properties and methods are inherited.  
- 🔄 **Child/Derived Class** – The new class that extends the functionality of the parent class.  
- 🔁 **Code Reusability** – Avoids redundancy by reusing common functionality.  
- 📚 **Hierarchical Organization** – Helps structure code efficiently.  

### 🔹 Example:  
```python
class Patient:
    def __init__(self, bmi, health_condition):
        self.bmi = bmi
        self.health_condition = health_condition

class EmergencyPatient(Patient):  # Inheriting from Patient
    def __init__(self, bmi, health_condition, severity_level):
        super().__init__(bmi, health_condition)
        self.severity_level = severity_level


````python
# Parent class: MedicalScan
class MedicalScan:
    def __init__(self, patient_name, scan_type, date, diagnosis):
        self.patient_name = patient_name
        self.scan_type = scan_type
        self.date = date
        self.diagnosis = diagnosis

    def display_scan_info(self):
        return f"Patient: {self.patient_name}, Scan: {self.scan_type}, Date: {self.date}, Diagnosis: {self.diagnosis}"


# Child class 1: MRI Scan
class MRI(MedicalScan):
    def __init__(self, patient_name, date, diagnosis, magnetic_field_strength):
        super().__init__(patient_name, "MRI", date, diagnosis)
        self.magnetic_field_strength = magnetic_field_strength  # Measured in Tesla

    def display_mri_info(self):
        return f"{self.display_scan_info()}, Magnetic Field: {self.magnetic_field_strength}T"


# Child class 2: PET Scan
class PET(MedicalScan):
    def __init__(self, patient_name, date, diagnosis, radiotracer):
        super().__init__(patient_name, "PET", date, diagnosis)
        self.radiotracer = radiotracer  # Type of radiotracer used

    def display_pet_info(self):
        return f"{self.display_scan_info()}, Radiotracer: {self.radiotracer}"


# Child class 3: Ultrasound Scan
class Ultrasound(MedicalScan):
    def __init__(self, patient_name, date, diagnosis, frequency):
        super().__init__(patient_name, "Ultrasound", date, diagnosis)
        self.frequency = frequency  # Measured in MHz

    def display_ultrasound_info(self):
        return f"{self.display_scan_info()}, Frequency: {self.frequency} MHz"


# Creating objects
scan1 = MRI("Alice Johnson", "2025-02-15", "Brain Tumor", 3.0)
scan2 = PET("Michael Smith", "2025-02-16", "Lung Cancer", "Fluorodeoxyglucose (FDG)")
scan3 = Ultrasound("Emma Davis", "2025-02-17", "Kidney Stones", 5.0)

# Printing scan details
print(scan1.display_mri_info())
print(scan2.display_pet_info())
print(scan3.display_ultrasound_info())
````

## 🔹 Explanation:

- 🏥 **MedicalScan (Parent Class)** – Represents general medical imaging scans.  
- 🧲 **MRI (Child Class 1)** – Adds a property for **magnetic field strength**.  
- 🔬 **PET (Child Class 2)** – Includes the **radiotracer** used in the scan.  
- 📡 **Ultrasound (Child Class 3)** – Specifies the **frequency** of the ultrasound waves.  

## 🏥 Medical Imaging Inheritance Example

In this example, we have a **parent class** `MedicalScan`, which represents general medical imaging scans. It includes a constructor (`__init__`) that initializes common properties such as `patient_name` and `scan_date`.  

### 🩺 Child Classes: Specializing Medical Scans  
We then define **three child classes**—`MRI`, `PET`, and `Ultrasound`—which inherit from the `MedicalScan` class.

- **🧲 MRI (Magnetic Resonance Imaging)**
  - Adds a `magnetic_field_strength` property to specify the power of the magnetic field used in the scan.

- **🧪 PET (Positron Emission Tomography)**
  - Introduces a `radiotracer` property, defining the substance used to highlight areas of metabolic activity.

- **📡 Ultrasound**
  - Adds a `frequency` property, representing the sound wave frequency used in imaging.

### 🔄 How Inheritance Works  
Each child class calls the parent class’s constructor using `super()`, ensuring that the general properties are initialized while also defining their unique attributes.  

When we create instances of these child classes, they inherit the properties and methods of the parent class. This allows us to access shared properties like `patient_name` and `scan_date`, while also utilizing the specialized attributes of each imaging type.


## 🔄 Polymorphism in Object-Oriented Programming  

Polymorphism is a fundamental concept in **object-oriented programming (OOP)** that allows objects of different classes to be treated as objects of a common **parent class**.  

This enables us to write flexible and reusable code that can work with **different types of objects in a uniform way**.  

### ✅ Why Polymorphism?  
- Allows **different classes** to share the **same method names** but have different implementations.  
- Promotes **code reusability** and **extensibility**.  
- Makes it easier to handle **different object types dynamically**.  

````python
# Parent class
class MedicalScan:
    def scan_type(self):
        return "General medical imaging scan"

# Child classes with specific implementations
class MRI(MedicalScan):
    def scan_type(self):
        return "Magnetic Resonance Imaging (MRI) - Uses magnetic fields"

class PET(MedicalScan):
    def scan_type(self):
        return "Positron Emission Tomography (PET) - Uses radiotracers"

class Ultrasound(MedicalScan):
    def scan_type(self):
        return "Ultrasound - Uses high-frequency sound waves"

# Function demonstrating polymorphism
def scan_details(scan):
    print(scan.scan_type())

# Creating objects of different scan types
mri_scan = MRI()
pet_scan = PET()
ultrasound_scan = Ultrasound()

# Calling the function with different scan types
scan_details(mri_scan)        # MRI-specific method
scan_details(pet_scan)        # PET-specific method
scan_details(ultrasound_scan) # Ultrasound-specific method
````

````python
import torch
from torch.utils.data import Dataset
import pandas as pd

class DiabetesDataset(Dataset):
    def __init__(self, csv_file):
        """
        Custom Dataset for loading diabetes data.
        Args:
        csv_file (str): Path to the CSV file containing the dataset.
        """
        self.data = pd.read_csv(csv_file)
        self.features = self.data.iloc[:, :-1].values  # All columns except the last one
        self.labels = self.data.iloc[:, -1].values     # Last column as labels

    def __len__(self):
        """Returns the total number of samples in the dataset."""
        return len(self.data)

    def __getitem__(self, idx):
        """
        Retrieves a single sample from the dataset.
        
        Args:
        idx (int): Index of the sample.

        Returns:
        tuple: (features, label) where features is a tensor of input data, and label is the corresponding target.
        """
        features = torch.tensor(self.features[idx], dtype=torch.float32)
        label = torch.tensor(self.labels[idx], dtype=torch.long)  # Assuming label is categorical (0 or 1)
        return features, label


dataset = DiabetesDataset("/kaggle/input/diabetes-data-set/diabetes.csv")
print(f"Dataset size: {len(dataset)}")
sample_features, sample_label = dataset[0]
print(f"Sample Features: {sample_features}")
print(f"Sample Label: {sample_label}")
````

## 🔍 Explanation:

- 🔹 **DiabetesDataset** class loads the dataset from a CSV file.  
- 🏗 **`__init__`** – Reads the dataset using **pandas**, extracts the **features** (input data) and **labels** (targets).  
- 🔄 **`__len__`** – Returns the **total number of samples** in the dataset. This helps PyTorch determine dataset size when using **DataLoader**.  
- 📤 **`__getitem__`** – Retrieves a **single sample** (features, label) as tensors when indexing the dataset.

````python
import os
import torch
from torch.utils.data import Dataset
from torchvision import transforms
from PIL import Image
````

````python
# Custom Dataset for Brain Tumor Images
class BrainTumorDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        """
        Args:
            root_dir (str): Path to dataset directory (either Training or Testing).
            transform (callable, optional): Optional transforms to apply to images.
        """
        self.root_dir = root_dir
        self.transform = transform
        self.classes = ["glioma", "meningioma", "notumor", "pituitary"]
        
        # Collect all image paths and corresponding labels
        self.image_paths = []
        self.labels = []
        
        for label, class_name in enumerate(self.classes):
            class_dir = os.path.join(root_dir, class_name)
            for img_name in os.listdir(class_dir):
                img_path = os.path.join(class_dir, img_name)
                self.image_paths.append(img_path)
                self.labels.append(label)
    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        image = Image.open(img_path).convert("RGB")  # Ensure 3-channel image
        label = self.labels[idx]

        if self.transform:
            image = self.transform(image)

        return image, label
````

## 📌 Explanation  

- 🏗 **BrainTumorDataset** class loads brain tumor images from **Training** and **Testing** folders.  
- 🔄 `__len__()` returns the dataset size.  
- 📤 `__getitem__()` loads an image, applies transformations, and returns `(image, label)`.  
- 🏷 **Labels:**  
  - `0 → Glioma`  
  - `1 → Meningioma`  
  - `2 → No Tumor`  
  - `3 → Pituitary`  
- 📌 **DataLoader** loads the dataset in **batches of 8** for training and testing.  

````python
# Define transformations (resize, convert to tensor)
transform = transforms.Compose([
    transforms.Resize((124, 124)),  # Resize to match model input size
    transforms.ToTensor()
])

# Create Dataset Instances
train_dataset = BrainTumorDataset(root_dir="/kaggle/input/brain-tumor-mri-dataset/Training", transform=transform)
test_dataset = BrainTumorDataset(root_dir="/kaggle/input/brain-tumor-mri-dataset/Testing", transform=transform)

# Create DataLoaders
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=8, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=8, shuffle=False)

# Iterate through the train DataLoader
for images, labels in train_loader:
    print("Batch of images shape:", images.shape)
    print("Batch of labels:", labels)
    break  # Print only one batch
````

````python
import torch
import torch.nn as nn

class BrainTumorCNN(nn.Module):
    def __init__(self, num_classes=4):
        super(BrainTumorCNN, self).__init__()

        # Convolutional layers
        self.conv1 = nn.Conv2d(in_channels=3, out_channels=32, kernel_size=3, stride=1, padding=1)
        self.relu1 = nn.ReLU()
        self.pool1 = nn.MaxPool2d(kernel_size=2, stride=2)

        self.conv2 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, stride=1, padding=1)
        self.relu2 = nn.ReLU()
        self.pool2 = nn.MaxPool2d(kernel_size=2, stride=2)

        # Fully connected layers
        self.flatten = nn.Flatten()  # Flattens for FC layer
        self.fc1 = nn.Linear(64 * 31 * 31, 128)
        self.relu3 = nn.ReLU()
        self.fc2 = nn.Linear(128, num_classes)  # Output layer

    def forward(self, x):
        x = self.conv1(x)
        x = self.relu1(x)
        x = self.pool1(x)

        x = self.conv2(x)
        x = self.relu2(x)
        x = self.pool2(x)

        x = self.flatten(x)  # Flatten before FC layer
        x = self.fc1(x)
        x = self.relu3(x)
        x = self.fc2(x)

        return x
````

````python
# Training loop function
def train_model(model, train_loader, criterion, optimizer, device, num_epochs=10):
    model.train()  # Set model to training mode

    for epoch in range(num_epochs):
        running_loss = 0.0
        correct = 0
        total = 0

        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)  # Move to GPU if available

            # Forward pass
            outputs = model(images)
            loss = criterion(outputs, labels)

            # Backward pass and optimization
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            running_loss += loss.item()

            # Calculate accuracy
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

        avg_loss = running_loss / len(train_loader)
        accuracy = 100 * correct / total

        print(f"Epoch [{epoch+1}/{num_epochs}] - Loss: {avg_loss:.4f}, Accuracy: {accuracy:.2f}%")

    print("Training complete! 🚀")
````

````python
# Train the model
train_model(model, train_loader, criterion, optimizer, device, num_epochs=2)
````
