# Area Plots 📈

An **Area Plot** (or stacked area plot) is used to visualize cumulative data over time. It is similar to a line plot, but the area below the line is filled with color, making it easier to see the contribution of each component to the whole.

For example; let's see the number of admissions by type, for each month :

````python
# Convert 'Entry Date' to datetime
df['Entry Date'] = pd.to_datetime(df['Entry Date'])

# Extract Month from 'Entry Date' to group by
df['Month'] = df['Entry Date'].dt.to_period('M')  # You can also use 'Y' for Year

# Group by 'Month' and 'Admission Type' to get the counts
admission_counts = df.groupby(['Month', 'Admission Type']).size().unstack(fill_value=0)

# Create an Area Plot
admission_counts.plot(kind='area', stacked=False)

# Add title, labels, and legend
plt.title("Admission Type Distribution Over Time")
plt.xlabel("Time (Month)")
plt.ylabel("Number of Admissions")
plt.legend(title="Admission Type", loc="upper left")

# Display the plot
plt.show()
````

As you see, it doesn't look good --> That's the effect of *stacked*

````python
# Extract Month from 'Entry Date' to group by
df['Month'] = df['Entry Date'].dt.to_period('M')  # You can also use 'Y' for Year

# Group by 'Month' and 'Admission Type' to get the counts
admission_counts = df.groupby(['Month', 'Admission Type']).size().unstack(fill_value=0)

# Create an Area Plot
admission_counts.plot(kind='area', stacked=True) # By changing Stacked to "True"

# Add title, labels, and legend
plt.title("Admission Type Distribution Over Time")
plt.xlabel("Time (Month)")
plt.ylabel("Number of Admissions")
plt.legend(title="Admission Type", loc="upper left")

# Display the plot
plt.show()
````

We can make th stacked plots transparent by changing *alpha* value

````python
# Extract Month from 'Entry Date' to group by
df['Month'] = df['Entry Date'].dt.to_period('M')  # You can also use 'Y' for Year

# Group by 'Month' and 'Admission Type' to get the counts
admission_counts = df.groupby(['Month', 'Admission Type']).size().unstack(fill_value=0)

# Create an Area Plot
admission_counts.plot(kind='area', stacked=True, alpha=0.5) # Play with alpha value to see the difference

# Add title, labels, and legend
plt.title("Admission Type Distribution Over Time")
plt.xlabel("Time (Month)")
plt.ylabel("Number of Admissions")
plt.legend(title="Admission Type", loc="upper left")

# Display the plot
plt.show()
````

#Bar Plots 📊

A bar plot is a way of representing data where the *length* of the bars represents the magnitude/size of the feature/variable. Bar graphs usually represent numerical and categorical variables grouped in intervals.

**Use case:** Let's see the disease distribution in our patients :

````python
# Count the number of patients for each disease
disease_counts = df['Disease'].value_counts()

# Create a bar plot (using a bar chart here as it is better suited for categorical data)
disease_counts.plot(kind='bar', edgecolor='black')

# Add title and labels
plt.title("Disease Distribution Among Patients")
plt.xlabel("Disease")
plt.ylabel("Number of Patients")

# Display the plot
plt.show()
````

Let's try the distribution by patient gender : answering the question : ***which gender has more disease?***

````python
# Now we Group the data by Gender and Disease, and count occurrences
gender_disease_counts = df.groupby(['Gender', 'Disease']).size().unstack()

# Plot the bar plot for disease distribution by gender
gender_disease_counts.plot(kind='bar', stacked=True, edgecolor='black')

# Add title and labels
plt.title("Disease Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Patients")

# Display the plot
plt.show()
````

Stacked this way in not much readable nor representable, so let's try to group by disease first :

````python
# Now we Group the data by Gender and Disease, and count occurrences
gender_disease_counts = df.groupby(['Disease', 'Gender']).size().unstack()

# Disease by Gender
gender_disease_counts.plot(kind='bar', edgecolor='black')

plt.title("Disease Distribution by Gender")
plt.xlabel("Disease")
plt.ylabel("Number of Patients")
plt.show()
````

Better ✔ we can also change the colors used for the bars, here :

````python
# Plotting
gender_disease_counts.plot(kind='bar', color=["#891652", "#EABE6C"])
plt.title('Count of Diseases by Gender')
plt.xlabel('Disease')
plt.ylabel('Number of Patients')
plt.xticks(rotation=0) #eliminate the name rotation
plt.legend(title='Gender')
plt.show()
````

- What we have been plotting is what's called Vertical bar plots, which are the default type of bar plots

Vertical bar plots are particuarly useful in analyzing time series data. One disadvantage is that they lack space for text labelling at the foot of each bar.

###Horizontal Bar Plot 📊

Sometimes it is more practical to represent the data horizontally, especially if you need more room for labelling the bars. In horizontal bar plots, the y-axis is used for labelling, and the length of bars on the x-axis corresponds to the magnitude of the variable being measured. As you will see, there is more room on the y-axis to  label categetorical variables.


For that we pass **barh** to kind 👇

````python
df.groupby('Disease')['Billing Amount'].mean().sort_values().plot(kind='barh', color='coral') # barh for horizontal
plt.title('Average Billing Amount per Disease')
plt.xlabel('Billing Amount')
plt.ylabel('Disease')
plt.show()
````

# Histograms 📉

A histogram is a way of representing the *frequency* distribution of numeric dataset.
##The way it works ⁉
it partitions the *x-axis* into ***bins***, assigns each data point in our dataset to a bin, and then counts the number of data points that have been assigned to each bin. So the *y-axis* is the frequency or the number of data points in each bin.

````python
# Plotting Age distribution
plt.hist(df['Age'], bins=10, edgecolor='black', color='skyblue')

plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
````

# Pie Charts 🥧

A `pie chart` is a circular graphic that displays numeric proportions by dividing a circle (or pie) into proportional slices. You are most likely already familiar with pie charts as it is widely used in business and media.
We can create pie charts in Matplotlib using `plt.pie()` OR by passing in the `kind=pie` keyword.

````python
df['Gender'].value_counts().plot(
    kind='pie',  #Type of the plot
    autopct='%1.1f%%', # Displays percentage values on the pie slices.
    startangle=90, # Rotates the starting position of the first slice to 90 degrees --> Try 0 to see the difference
    figsize=(5, 6), # Figure size
    explode = (0.05, 0) , # Explode the 1st slice
    shadow=True, # Makes a shadow
    colors=['#ff9999', '#66b3ff'] # The colors to use
)
plt.title("Patient Gender Distribution")
plt.ylabel("")
plt.show()
````

# Scatter Plots <a id="10"></a>

A `scatter plot` (2D) is a useful method of comparing variables against each other. `Scatter` plots look similar to `line plots` in that they both map independent and dependent variables on a 2D graph.

While the datapoints are connected together by a line in a line plot, they are not connected in a scatter plot. The data in a scatter plot is considered to express a trend.

- With further analysis using tools like regression, we can mathematically calculate this relationship and use it to predict trends outside the dataset.

````python
plt.scatter(df['Age'], df['Billing Amount'], alpha=0.6, color='purple')
plt.title('Age vs Billing Amount')
plt.xlabel('Age')
plt.ylabel('Billing Amount')
plt.show()
````

#Box Plot

A box plot (also known as a *box-and-whisker* plot) is a statistical graphic used to visualize the distribution of a dataset. It shows the central ***tendency***, ***variability***, and ***outliers*** of the data. A box plot is particularly useful for comparing distributions across different groups or categories.

````python
df['Stay Duration'] = (df['Discharge Date'] - df['Entry Date']).dt.days

# Group by Medication and calculate the distribution of Stay Duration
plt.figure(figsize=(12, 6))

df.boxplot(column='Stay Duration', by='Medication')

# Titles and labels
plt.title('Stay Duration by Medication')
plt.suptitle('')  # Remove the default title generated by boxplot
plt.xlabel('Stay Duration (days)')
plt.ylabel('Medication')

# Display the plot
plt.show()
````

Let's do some format changing to make it look better :

````python
df['Stay Duration'] = (df['Discharge Date'] - df['Entry Date']).dt.days

# Group by Medication and calculate the distribution of Stay Duration
plt.figure(figsize=(12, 6))
df.boxplot(column='Stay Duration', by='Medication',
           vert=False,  # Turn into vertical orientation
           patch_artist=True, # Fills the boxes with color.
           medianprops=dict(color="red", linewidth=2), # Customize the appearance of the median line inside the box.
           boxprops=dict(facecolor='lightblue', color='blue'), # Controls the appearance of the boxes in the box plot.
           whiskerprops=dict(color="blue", linewidth=1.5)) # Customize the appearance of the whiskers (the lines extending from the box).

# Titles and labels
plt.title('Stay Duration by Medication')
plt.suptitle('')  # Remove the default title generated by boxplot
plt.xlabel('Stay Duration (days)')
plt.ylabel('Medication')

# Display the plot
plt.show()
````
