## 2. Pandas Series — A Single Column of Data

A **Series** is like a single column in a spreadsheet. It has:
- An **index** (labels on the left)
- **Values** (the actual data on the right)
- A **name** (optional)

Think of it as a Python list, but with labels attached to each item.

````python
# Create a Series of tumor sizes (in cm) for 5 patients
tumor_sizes = pd.Series(
    [2.3, 1.8, 4.1, 0.9, 3.5],
    name="tumor_cm"
)
print(tumor_sizes)
print()
print(f"Type: {type(tumor_sizes)}")
````

````python
# Create a Series with custom index labels
tumor_sizes_named = pd.Series(
    [2.3, 1.8, 4.1, 0.9, 3.5],
    index=["Fatima", "Chidi", "Amina", "Bola", "Kofi"],
    name="tumor_cm"
)
print(tumor_sizes_named)
````

````python
# Access values by index
print(f"Fatima's tumor: {tumor_sizes_named['Fatima']} cm")
print(f"Largest tumor: {tumor_sizes_named.max()} cm")
print(f"Average tumor: {tumor_sizes_named.mean():.2f} cm")
````

## 3. Pandas DataFrame — A Table of Data

A **DataFrame** is like a full spreadsheet or database table:
- Multiple columns (each column is a Series)
- Rows and columns
- Different data types in different columns

This is the object you will work with 95% of the time in Pandas.

````python
# Create a DataFrame of patient records
# We will use this dataset throughout the tutorial
patients = pd.DataFrame({
    "name": ["Fatima", "Chidi", "Amina", "Bola", "Kofi"],
    "age": [52, 34, 45, 28, 61],
    "scan_type": ["MRI", "CT", "MRI", "X-Ray", "CT"],
    "tumor_cm": [3.5, 1.2, 4.1, 0.8, 2.3],
    "diagnosis": ["Glioma", "Normal", "Meningioma", "Normal", "Glioma"]
})

print(patients)
````

````python
# Each column is a Series
print(type(patients["name"]))
print()

# The whole table is a DataFrame
print(type(patients))
````

### Series vs DataFrame — Key Difference

| Feature | Series | DataFrame |
|---------|--------|-----------|
| Dimensions | 1D (one column) | 2D (rows × columns) |
| Think of it as | A single list with labels | A spreadsheet / table |
| Access | `df["tumor_cm"]` | `df[["name", "tumor_cm"]]` |
| Type | `pandas.Series` | `pandas.DataFrame` |

**Key insight:** A DataFrame is just a collection of Series glued together. Pull out one column → you get a Series. Combine Series → you get a DataFrame.

## 6. Selecting Columns and Rows

### Selecting Columns
- **Single brackets** `df["col"]` → returns a **Series**
- **Double brackets** `df[["col1", "col2"]]` → returns a **DataFrame**

````python
# Single column → Series
names = df["name"]
print(type(names))
print(names.head())
````

````python
# Multiple columns → DataFrame
subset = df[["name", "tumor_cm", "diagnosis"]]
print(type(subset))
print(subset.head())
````

### Selecting Rows with `iloc` and `loc`

| Method | How it works | Slicing behavior |
|--------|-------------|-----------------|
| `iloc` | By integer position (like list indexing) | Excludes end: `iloc[0:3]` → rows 0, 1, 2 |
| `loc` | By label/index | Includes end: `loc[0:3]` → rows 0, 1, 2, 3 |

````python
# iloc — by position (like Python lists)
print("First patient:")
print(df.iloc[0])
print()

print("First 3 patients:")
print(df.iloc[0:3])
````

````python
# loc — by label (includes the end!)
print("Patient at index 5, name and diagnosis only:")
print(df.loc[5, ["name", "diagnosis"]])
print()

print("Patients 0 to 2, specific columns:")
print(df.loc[0:2, ["name", "age", "tumor_cm"]])
````

## 7. Filtering Data — Boolean Indexing

This is one of the most powerful features in Pandas. You write a condition, and Pandas returns only the rows that match.

**Syntax:** `df[condition]`

````python
# Patients with tumor larger than 3.0 cm
critical = df[df["tumor_cm"] > 3.0]
print(f"Critical patients (tumor > 3.0 cm): {len(critical)}")
print(critical[["name", "tumor_cm", "diagnosis"]])
````

````python
# Only MRI scans
mri_patients = df[df["scan_type"] == "MRI"]
print(f"MRI patients: {len(mri_patients)}")
print(mri_patients[["name", "scan_type", "diagnosis"]])
````

````python
# Multiple conditions — use & for AND, | for OR
# IMPORTANT: wrap each condition in parentheses!

# Glioma patients with tumor > 2.0 cm
glioma_large = df[
    (df["diagnosis"] == "Glioma") & (df["tumor_cm"] > 2.0)
]
print("Glioma patients with tumor > 2.0 cm:")
print(glioma_large[["name", "tumor_cm", "diagnosis"]])
````

````python
# Female patients from Nigeria with MRI
nigerian_women_mri = df[
    (df["gender"] == "F") &
    (df["country"] == "Nigeria") &
    (df["scan_type"] == "MRI")
]
print("Nigerian female MRI patients:")
print(nigerian_women_mri[["name", "country", "scan_type", "diagnosis"]])
````

````python
# Using .isin() for multiple values
# Patients with Glioma OR Meningioma
tumor_patients = df[df["diagnosis"].isin(["Glioma", "Meningioma"])]
print(f"Tumor patients (Glioma or Meningioma): {len(tumor_patients)}")
print(tumor_patients[["name", "diagnosis", "tumor_cm"]])
````

## 8. Data Cleaning — Missing Values

Real hospital data always has missing values. A patient's age was not recorded, a tumor measurement was not taken, a scan date was left blank.

In Pandas, missing values appear as **NaN** (Not a Number).

````python
# Check for missing values in each column
print("Missing values per column:")
print(df.isnull().sum())
print()
print(f"Total missing values: {df.isnull().sum().sum()}")
````

````python
# Which rows have missing ages?
missing_age = df[df["age"].isnull()]
print("Patients with missing age:")
print(missing_age[["patient_id", "name", "age", "country"]])
````

### Handling Missing Values — Drop or Fill?

| Strategy | When to use | Code |
|----------|------------|------|
| **Drop rows** | Few missing (<5%), data is not critical | `df.dropna()` |
| **Fill with median** | Numeric columns, cannot lose rows | `df.fillna(df.median())` |
| **Fill with mode** | Categorical columns | `df.fillna(df.mode()[0])` |
| **Fill with value** | You know the default | `df.fillna(0)` |

````python
# Option 1: Fill missing ages with the median
median_age = df["age"].median()
print(f"Median age: {median_age}")

df["age"] = df["age"].fillna(median_age)
print(f"Missing ages after fill: {df['age'].isnull().sum()}")
````

````python
# Option 2: Drop rows where tumor_cm is missing
# (tumor measurement is critical — we cannot guess it)
print(f"Shape before dropping: {df.shape}")
df = df.dropna(subset=["tumor_cm"])
print(f"Shape after dropping rows with missing tumor_cm: {df.shape}")
````

````python
# Verify — no more missing values
print("Missing values after cleaning:")
print(df.isnull().sum())
````

## 11. Adding & Modifying Columns

Create new columns from existing data — this is called **feature engineering**.

````python
# Add a status column based on tumor size
df["status"] = df["tumor_cm"].apply(
    lambda x: "Critical" if x > 3.0 else ("Monitor" if x > 1.5 else "Stable")
)
print(df[["name", "tumor_cm", "status"]].head(10))
````

````python
# Math on columns — convert cm to mm
df["tumor_mm"] = df["tumor_cm"] * 10
print(df[["name", "tumor_cm", "tumor_mm"]].head())
````

````python
# Using np.where for simple binary conditions
df["is_critical"] = np.where(df["tumor_cm"] > 3.0, True, False)
print(df[["name", "tumor_cm", "is_critical"]].head(10))
````

````python
# String operations — standardize text
df["name_upper"] = df["name"].str.upper()
df["diagnosis_lower"] = df["diagnosis"].str.lower()
print(df[["name", "name_upper", "diagnosis", "diagnosis_lower"]].head())
````

## 13. Descriptive Statistics

Pandas makes statistical analysis incredibly easy. One line of code gives you a complete statistical profile.

````python
# Full statistical summary of numeric columns
df.describe()
````

| Statistic | Meaning |
|-----------|---------|

## 14. GroupBy — Split, Apply, Combine

`groupby()` splits your data into groups, applies a function to each group, and combines the results. It is one of Pandas' most powerful features.

Think of it as: *"For each diagnosis, what is the average tumor size?"*

````python
# Average tumor size by diagnosis
print("=== Average tumor size by diagnosis ===")
print(df.groupby("diagnosis")["tumor_cm"].mean())
````

````python
# Multiple statistics by diagnosis
print("=== Tumor stats by diagnosis ===")
print(df.groupby("diagnosis")["tumor_cm"].describe())
````

````python
# Multiple aggregations with .agg()
print("=== Country-level summary ===")
country_summary = df.groupby("country").agg({
    "patient_id": "count",
    "age": "mean",
    "tumor_cm": ["mean", "max"]
})
print(country_summary)
````

````python
# Count of each scan type per country
print("=== Scan types by country ===")
print(pd.crosstab(df["country"], df["scan_type"]))
````

## 15. Correlation Analysis

Correlation tells you if two numeric variables move together.

| Value | Meaning |
|-------|---------|
| +1.0 | Perfect positive (both go up together) |
| 0.0 | No relationship |
| -1.0 | Perfect negative (one goes up, other goes down) |

````python
# Correlation between age and tumor size
corr = df["age"].corr(df["tumor_cm"])
print(f"Correlation between age and tumor size: {corr:.3f}")

if abs(corr) > 0.5:
    direction = "positive" if corr > 0 else "negative"
    print(f"→ Strong {direction} correlation: older patients tend to have {'larger' if corr > 0 else 'smaller'} tumors")
elif abs(corr) > 0.3:
    print("→ Moderate correlation")
else:
    print("→ Weak or no correlation")
````

````python
# Correlation matrix for all numeric columns
print("=== Correlation Matrix ===")
numeric_cols = df.select_dtypes(include=[np.number])
print(numeric_cols.corr().round(3))
````

## 16. The Complete Data Analysis Workflow

Every data project follows these 5 steps. Memorize them:

**Load → Explore → Clean → Analyze → Save**

````python
# ============================
# STEP 1: LOAD
# ============================
raw = pd.read_csv("hospital_patients.csv")
print("Step 1: Data loaded")
print(f"  Shape: {raw.shape}")

# ============================
# STEP 2: EXPLORE
# ============================
print("\nStep 2: Exploring data")
print(f"  Columns: {raw.columns.tolist()}")
print(f"  Missing values:\n{raw.isnull().sum()}")
print(f"  Data types:\n{raw.dtypes}")

# ============================
# STEP 3: CLEAN
# ============================
clean = raw.copy()  # Always work on a copy

# Fill missing ages with median
clean["age"] = clean["age"].fillna(clean["age"].median())

# Drop rows with missing tumor measurements
clean = clean.dropna(subset=["tumor_cm"])

# Convert date column
clean["scan_date"] = pd.to_datetime(clean["scan_date"])

# Remove duplicates
clean = clean.drop_duplicates()

print(f"\nStep 3: Data cleaned")
print(f"  Shape after cleaning: {clean.shape}")
print(f"  Missing values: {clean.isnull().sum().sum()}")

# ============================
# STEP 4: ANALYZE
# ============================
print("\nStep 4: Analysis results")
print(f"  Average age: {clean['age'].mean():.1f}")
print(f"  Average tumor: {clean['tumor_cm'].mean():.2f} cm")
print(f"\n  Diagnosis distribution:")
print(clean["diagnosis"].value_counts().to_string())
print(f"\n  Average tumor by diagnosis:")
print(clean.groupby("diagnosis")["tumor_cm"].mean().round(2).to_string())

# ============================
# STEP 5: SAVE
# ============================
clean.to_csv("cleaned_patients.csv", index=False)

# Save just critical patients
critical = clean[clean["tumor_cm"] > 3.0]
critical.to_csv("critical_patients.csv", index=False)

print(f"\nStep 5: Files saved")
print(f"  cleaned_patients.csv ({clean.shape[0]} patients)")
print(f"  critical_patients.csv ({critical.shape[0]} critical patients)")

print("\n✅ Complete workflow done!")
````

## 17. Bonus: String Operations

Hospital data is full of text inconsistencies: "MRI", "mri", " MRI ", "Mri Brain". The `.str` accessor lets you clean all of that.

````python
# Common string operations
example_df = pd.DataFrame({
    "diagnosis_raw": ["  Glioma ", "glioma", "GLIOMA", "Glioma "],
    "scan_raw": ["MRI Brain", "CT Abdomen", "X-Ray Chest", "MRI Spine"]
})

# Standardize: lowercase + strip whitespace
example_df["diagnosis_clean"] = example_df["diagnosis_raw"].str.lower().str.strip()
print("Cleaned diagnoses:")
print(example_df[["diagnosis_raw", "diagnosis_clean"]])
````

## 18. Common Pandas Errors (and How to Fix Them)

You **will** see these errors. Do not panic — they all have simple fixes.

````python
# ERROR 1: KeyError — column name does not exist
try:
    df["tumor"]  # Wrong! The column is "tumor_cm"
except KeyError as e:
    print(f"❌ KeyError: {e}")
    print("   Fix: Check df.columns for the exact column name")
    print(f"   Available columns: {df.columns.tolist()}")
````

````python
# ERROR 2: FileNotFoundError — wrong file path
import os

filepath = "nonexistent_file.csv"
if os.path.exists(filepath):
    df_test = pd.read_csv(filepath)
else:
    print(f"❌ File not found: {filepath}")
    print("   Fix: Use os.path.exists() to check first")
    print("   Fix: Check your file path and spelling")
````

````python
# ERROR 3: ValueError — trying to convert bad data
bad_data = pd.Series(["10", "twenty", "30", "N/A", "50"])
print("Bad data:", bad_data.tolist())

# This would crash: bad_data.astype(int)
# Instead, use to_numeric with errors='coerce'
good_data = pd.to_numeric(bad_data, errors="coerce")
print("After to_numeric(errors='coerce'):")
print(good_data)
print("   Non-numeric values became NaN — safe!")
````
